#### **Desafio - Estruturar uma Função para Limpeza da Tabela Importada**
---

Desenvolver uma função para a limpeza e tratamento dos dados, retornando duas tabelas 'população' e 'unida'.

---

Para a resolução desse desafio, considerei que tinha apenas a *URL* da página de onde seriam retiradas as informações e o *DataFrame* com os dados de gastos do mês mais recente.

Optei por trabalhar com funções aninhadas, em vez de trabalhar com uma única função, principalmente para facilitar o entendimento e praticar a chamada de funções.

````python
display(atends_mes_recente)
display(url_wiki)
````

````python
def importa_wiki(url):
    response=r.get(url)

    soup=BeautifulSoup(response.text, 'html.parser')
    tabela_pop=soup.find('table',{'class':"wikitable"})

    df_populacao=pd.DataFrame(pd.read_html(str(tabela_pop))[0])

    return df_populacao

def limpa_pop(suja):
    suja = suja.rename(columns={'Unidade federativa':'UF'})
    suja = suja.dropna()
    #suja['UF'] = suja['Unidade federativa']
    suja['População'] = suja['População'].str.replace(' ','').astype(int)
    suja['UF'] = suja['UF'].str.strip()

    limpa = suja[['UF','População']].set_index('UF')
    #limpa.index=limpa.index.str.strip()

    return limpa

def faz_tudo(url_w, tab_gastos):
    pop_suja = importa_wiki(url_w)
    pop_limpa = limpa_pop(pop_suja)
    result = pop_limpa.join(tab_gastos)

    return pop_limpa,result
````

````python
populacao, unida = faz_tudo(url_wiki,atends_mes_recente)
display(populacao)
display(unida)
````

````python
def insere_mes_tabela (tab_origem: pd.DataFrame, tab_destino: pd.DataFrame, mes: str):
    gastos_mes=tab_origem[mes]
    tab_destino[f'G_{mes}']=gastos_mes
    tab_destino[f'GpC_{mes}']=tab_destino[f'G_{mes}']/tab_destino['População']
````

Como achei o desafio relativamente fácil, resolvi "brincar" um pouco com os parâmetros de limites, marcação nos eixos **x** e **y**, *grid* e os títulos dos eixos.

A função **```melt```** faz o que eu fiz manualmente no Módulo 01. Ela converte as informações que estão nas colunas em dados na nossa tabela de dados.

No caso da base de dados que estamos usando agora, com os dados mensais em linhas com as 27 UFs em colunas diferentes, ao aplicar a função **```melt```** teremos cada linha virando 27 linhas e a informação da UF passa a ser uma coluna.

````python
dados_mensais_abertos = dados_mensais.reset_index().melt(id_vars = 'index',
                                 value_vars = dados_mensais.columns)
dados_mensais_abertos.columns = ['Mês','UF','Valor']
dados_mensais_abertos.head()
````

O primeiro que escolhi foi o **```strptime```**, que serve para transformar uma data *string* em um objeto do tipo *datetime* com seus diversos componentes, de acordo com o mapeamento que é passado no 2º parâmetro. Acabei olhando também o **```strftime```**, que faz o caminho inverso.

Será muito útil no tratamento de importações de dados de diferentes fontes, com diferentes formatações.

Os possíveis parâmetros são:

```%a``` - Dia da Semana abreviado - Sun, Mon, …, Sat (en_US);

```%A``` - Dia da Semana por extenso - Sunday, Monday, …, Saturday (en_US);

```%w``` - Dia da Semana numérico - 0 = Domingo, 1 = 2ª-feira, ...

```%d``` - Dia do Mês com 2 dígitos - 01 a 31

```%b``` - Mês abreviado - Jan, Feb, …, Dec (en_US);

```%B``` - Mês por extenso - January, February, …, December (en_US);

```%m``` - Mês numérico com 2 dígitos - 01 a 12;

```%y``` - Ano com 2 dígitos - 00 a 99;

```%Y``` - Ano com 4 dígitos - 0001 a 9999;

```%H``` - Hora de 24 horas com 2 dígitos - 00 a 23;

```%I``` - Hora de 12 horas com 2 dígitos - 00 a 12;

```%p``` - Indicação de AM/PM - AM, PM (en_US);

```%M``` - Minuto com 2 dígitos- 00, 01, …, 59

```%S``` - Segundo com 2 dígitos - 00, 01, …, 59

```%f``` - Microssegundo com 6 dígitos - 000000, 000001, …, 999999

```%z``` - Diferença de fuso horário no formato ±HHMM[SS[.ffffff]] - +0000, -0400, +1030, +063415, -030712.345216

```%Z``` - Nome da Zona de Fuso - (vazio), UTC, GMT

```%j``` - Dia do Ano com 3 dígitos - 001, 002, …, 366

```%U``` - Número da Semana com 2 dígitos, Semana inicia no Domingo e a semana 0 do ano são os dias anteriores ao primeiro Domingo - 00, 01, …, 53

```%W``` - Número da Semana com 2 dígitos, Semana inicia no 2ª-feira e a semana 0 do ano são os dias anteriores à primeira 2ª-feira - 00, 01, …, 53

```%c``` - Formato local de apresentação de data e hora - Tue Aug 16 21:30:00 1988 (en_US);

```%x``` - Formato local de apresentação de data - 08/16/1988 (en_US);

```%X``` - Formato local de apresentação de hora - 21:30:00 (en_US);

A comparação dos meses nos mostra uma queda, todos os anos, no mês de fevereiro. Mas isso pode ser causado pelo menor número de dias que temos nesse mês.

Vamos então calcular o valor médio diário gasto, para verificar esse efeito. Para isso, vamos criar um dicionário com os números de dias que temos em cada mês e aplicar em nossa base de dados.

````python
dias_por_mes = {
    1  : 31, 2  : 28, 3  : 31,
    4  : 30, 5  : 31, 6  : 30,
    7  : 31, 8  : 31, 9  : 30,
    10 : 31, 11 : 30, 12 : 31
}
````

#### **DESAFIO EXTRA - Criar uma função que calcule efetivamente o número de dias de cada mês, para funcionar corretamente em anos bissextos**

````python
def mapeia_dias(dia_qualquer: datetime):
    '''Mapeia a quantidade de dias que tem no mês
       Recebe uma data qualquer (tipo datetime) e
       retorna o número de dias do mês'''

    ano = dia_qualquer.timetuple()[0]
    mês = dia_qualquer.timetuple()[1]
    primeiro, dias = calendar.monthrange(ano, mês)
    return dias
````

### **Desafio - Reordenar os Tons de Cor pelo Eixo Y**
---

O tom da cor passa a ser ligado ao eixo **Y** e não ao eixo **X**, que são os anos, como está até agora. Caso contrário, a variação do tom apenas reforça a mesma informação que já temos, de que quanto mais à direita maiores são os anos representados.

````python
### DESAFIO
ordem_valores = dados_anuais_uf_analisar['Valor'].rank(method='max').astype(int) - 1

cores = sns.color_palette('flare',len(dados_anuais_uf_analisar.index))
paleta_cores = [cores[x] for x in ordem_valores]

sns.barplot(data=dados_anuais_uf_analisar,
            x=dados_anuais_uf_analisar.index,
            y='Valor',
            palette = paleta_cores)
````

No caso do exemplo acima temos apenas uma inversão de sequência (2015 e 2016), mas em séries com maior alternância de valores essa vinculação da tonalidade ao valor se torna mais interessante.

Vamos agora visualizar mais estados em um mesmo gráfico e ver que ferramentas podemos usar para melhorar a visualização.

Começamos com a separação de 3 UFs para a geração do *DataFrame* que iremos usar nessa análise. É importante perceber que agora o `groupby` passa a ter como argumento uma lista das colunas que serão as quebras de agrupamento, no caso, além do 'Ano', a própria 'UF'.

````python
lista_ufs = ['Rio Grande do Sul', 'Santa Catarina', 'Paraná']
dados_anuais_lista_ufs = dados_mensais_abertos.query('UF in @lista_ufs').groupby(['UF', 'Ano']).sum().reset_index()
dados_anuais_lista_ufs.head()
````

Vamos plotar um gráfico de barras simples, para ver o que iremos obter.

````python
sns.barplot(data=dados_anuais_lista_ufs,
                x='Ano',
                y='Valor')
````

Sem passar nenhum tipo de quebra para representar as UFs, o **`Seaborn`** vai plotar, em cada barra, uma linha central que representa o intervalo de confiança dos valores encontrados.

No caso atual, é uma informação de difícil utilização e análise. Podemos usar, como já fizemos antes, a diferença de tonalidade (`hue`) para indicar as diferentes UFs.

````python
sns.barplot(data=dados_anuais_lista_ufs,
            x='Ano',
            y='Valor',
            hue='UF')
````

Mas como vimos anteriormente, estamos colocando informações categóricas (as diferentes UFs) misturadas em meio às barras. Isso permite uma comparação entre as diferentes UFs em um mesmo ano, mas dificulta a variação da evolução dos valores ao longo do tempo.

Vamos usar então o `catplot`.

````python
xx = sns.catplot(data = dados_anuais_lista_ufs,
                 x = 'Ano',
                 y = 'Valor',
                 kind = 'bar',
                 col = 'UF',
                 palette = 'flare')

xx.set_axis_labels('Ano', 'Valores')

xx.set_xticklabels(rotation=30,ha ='right')
````

Na documentação da API do **```Seaborn```** ([link](https://seaborn.pydata.org/api.html)) temos os agrupamentos de gráficos que podem ser gerados como no exemplo acima, em que a própria biblioteca trata de gerar uma figura que comporte os diversos gráficos que serão gerados.

No caso acima usamos o ```catplot``` que agrupa os gráficos de categorias, mas também temos outros tipos de agrupamento que podem ser feitos.

Quando temos a necessidade de colocar gráficos diferentes na mesma figura, é melhor usar o **```matplotlib```**, que dá mais controle sobre os detalhes.

No caso acima, por exemplo, se quisermos colocar as cores variando de acordo com o eixo **Y** vamos precisar estabelecer uma sequência de cores para cada gráfico, não sendo possível trabalhar com uma paleta única.

---

Além de outras bibliotecas gráficas como **```Bokeh```** e **```pyplot```**, por exemplo.


**`matplotlib`**

É uma biblioteca voltada para a geração de gráficos com maior poder de customização para o usuário. Isso permite soluções mais complexas, com a passagem de cada um dos parâmetros que irão construir o gráfico a ser plotado.

Essa flexibilidade demanda maior quantidade de código, geralmente.

([Documentação](https://matplotlib.org/stable/contents.html))

---

**`pyplot`**

É um conjunto de comandos dentro do próprio **`matplotlib`**, criado com a intenção de deixá-lo parecido com o ***MatLab***, para facilitar a integração de quem era habituado, além de ser gratuito.

---

**`Seaborn`**

É uma biblioteca montada em cima do **`matplotlib`**. A ideia é simplificar a geração de gráficos sem perder a possibilidade de usar elementos configuráveis do **`matplotlib`**.

Apresenta uma documentação bem completa, não se limitando a explicar sintaxe de comandos e seus parâmetros. Nela é possível encontrar também tutoriais de como criar melhores gráficos inclusive auxiliando na escolha do tipo de gráfico e da paleta de cores a ser usada dependendo do tipo de dados que serão analisados.

([Tutoriais](https://seaborn.pydata.org/tutorial.html))

---

**`Bokeh`**

É uma biblioteca voltada para a criação de gráficos interativos que podem ser vistos e manipulados diretamente no *browser*.

É muito útil também em fase de investigação de dados, pois permite localizar *outliers* passando o ponteiro do *mouse* sobre pontos específicos, ampliar algum ponto a ser visualizado com mais atenção, etc.
