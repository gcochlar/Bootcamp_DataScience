#### **Desafio - Estruturar uma Função para Limpeza da Tabela Importada**
---

Desenvolver uma função para a limpeza e tratamento dos dados, retornando duas tabelas 'população' e 'unida'.

---

Para a resolução desse desafio, considerei que tinha apenas a *URL* da página de onde seriam retiradas as informações e o *DataFrame* com os dados de gastos do mês mais recente.

Optei por trabalhar com funções aninhadas, em vez de trabalhar com uma única função, principalmente para facilitar o entendimento e praticar a chamada de funções.

````python
display(atends_mes_recente)
display(url_wiki)
````

````python
def importa_wiki(url):
    response=r.get(url)

    soup=BeautifulSoup(response.text, 'html.parser')
    tabela_pop=soup.find('table',{'class':"wikitable"})

    df_populacao=pd.DataFrame(pd.read_html(str(tabela_pop))[0])

    return df_populacao

def limpa_pop(suja):
    suja = suja.rename(columns={'Unidade federativa':'UF'})
    suja = suja.dropna()
    #suja['UF'] = suja['Unidade federativa']
    suja['População'] = suja['População'].str.replace(' ','').astype(int)
    suja['UF'] = suja['UF'].str.strip()

    limpa = suja[['UF','População']].set_index('UF')
    #limpa.index=limpa.index.str.strip()

    return limpa

def faz_tudo(url_w, tab_gastos):
    pop_suja = importa_wiki(url_w)
    pop_limpa = limpa_pop(pop_suja)
    result = pop_limpa.join(tab_gastos)

    return pop_limpa,result
````

````python
populacao, unida = faz_tudo(url_wiki,atends_mes_recente)
display(populacao)
display(unida)
````

````python
def insere_mes_tabela (tab_origem: pd.DataFrame, tab_destino: pd.DataFrame, mes: str):
    gastos_mes=tab_origem[mes]
    tab_destino[f'G_{mes}']=gastos_mes
    tab_destino[f'GpC_{mes}']=tab_destino[f'G_{mes}']/tab_destino['População']
````

Como achei o desafio relativamente fácil, resolvi "brincar" um pouco com os parâmetros de limites, marcação nos eixos **x** e **y**, *grid* e os títulos dos eixos.

### Aula 04 - Vídeo 02

A função **```melt```** faz o que eu fiz manualmente no Módulo 01. Ela converte as informações que estão nas colunas em dados na nossa tabela de dados.

Isso vai reduzir drasticamente a quantidade de colunas e aumentar significativamente a quantidade de linhas.

No caso da base de dados que estamos usando agora, com os dados mensais em linhas com as 27 UFs em colunas diferentes, ao aplicar a função **```melt```** teremos cada linha virando 27 linhas e a informação da UF passa a ser uma coluna.

````python
dados_mensais_abertos = dados_mensais.reset_index().melt(id_vars = 'index',
                                 value_vars = dados_mensais.columns)
dados_mensais_abertos.columns = ['Mês','UF','Valor']
dados_mensais_abertos.head()
````

O primeiro que escolhi foi o **```strptime```**, que serve para transformar uma data *string* em um objeto do tipo *datetime* com seus diversos componentes, de acordo com o mapeamento que é passado no 2º parâmetro. Acabei olhando também o **```strftime```**, que faz o caminho inverso.

Será muito útil no tratamento de importações de dados de diferentes fontes, com diferentes formatações.

Os possíveis parâmetros são:

```%a``` - Dia da Semana abreviado - Sun, Mon, …, Sat (en_US);

```%A``` - Dia da Semana por extenso - Sunday, Monday, …, Saturday (en_US);

```%w``` - Dia da Semana numérico - 0 = Domingo, 1 = 2ª-feira, ...

```%d``` - Dia do Mês com 2 dígitos - 01 a 31

```%b``` - Mês abreviado - Jan, Feb, …, Dec (en_US);

```%B``` - Mês por extenso - January, February, …, December (en_US);

```%m``` - Mês numérico com 2 dígitos - 01 a 12;

```%y``` - Ano com 2 dígitos - 00 a 99;

```%Y``` - Ano com 4 dígitos - 0001 a 9999;

```%H``` - Hora de 24 horas com 2 dígitos - 00 a 23;

```%I``` - Hora de 12 horas com 2 dígitos - 00 a 12;

```%p``` - Indicação de AM/PM - AM, PM (en_US);

```%M``` - Minuto com 2 dígitos- 00, 01, …, 59

```%S``` - Segundo com 2 dígitos - 00, 01, …, 59

```%f``` - Microssegundo com 6 dígitos - 000000, 000001, …, 999999

```%z``` - Diferença de fuso horário no formato ±HHMM[SS[.ffffff]] - +0000, -0400, +1030, +063415, -030712.345216

```%Z``` - Nome da Zona de Fuso - (vazio), UTC, GMT

```%j``` - Dia do Ano com 3 dígitos - 001, 002, …, 366

```%U``` - Número da Semana com 2 dígitos, Semana inicia no Domingo e a semana 0 do ano são os dias anteriores ao primeiro Domingo - 00, 01, …, 53

```%W``` - Número da Semana com 2 dígitos, Semana inicia no 2ª-feira e a semana 0 do ano são os dias anteriores à primeira 2ª-feira - 00, 01, …, 53

```%c``` - Formato local de apresentação de data e hora - Tue Aug 16 21:30:00 1988 (en_US);

```%x``` - Formato local de apresentação de data - 08/16/1988 (en_US);

```%X``` - Formato local de apresentação de hora - 21:30:00 (en_US);


````python
# Usando datetime.strptime()
dt = datetime.datetime.strptime("05/06/2021 14:15", "%d/%m/%Y %H:%M")
print(dt)
print(dt.isoformat(sep = '!',timespec='microseconds'))
print(dt.isoformat(sep = '$',timespec='minutes'))
print(dt.isoformat(sep = '%',timespec='hours'))
````

Depois disso, resolvi estudar um pouco como o ***Python*** trata os cálculos de diferenças entre datas e horas.

Descobri que ao fazer a conta de diferença entre 2 datas, o retorno é em dias, horas, minutos, segundos e microssegundos (se for o caso). É necessário tratar de alguma forma os anos bissextos, se a intenção é calcular uma data de aniversário, por exemplo.

````python
from datetime import timedelta
````

````python
dt01 = datetime.datetime.strptime("26/03/1973 09:30", "%d/%m/%Y %H:%M")
dt02 = datetime.datetime.strptime("26/05/1983 19:35:30", "%d/%m/%Y %H:%M:%S")

x = dt02 - dt01
print(x)
````

````python
dt03 = dt01+datetime.timedelta(days=48*365)
dt03
````

A saída do ```datetime.datetime``` vai ser sempre no formato (ano, mês, dia, hora, minuto, segundo, microssegundo), desde que haja todos os componentes.
