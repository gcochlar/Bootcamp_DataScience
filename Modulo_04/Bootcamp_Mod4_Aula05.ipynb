#### **Desafio - Pesquisar Aprendizado Supervisionado, Não Supervisionado, Por Reforço e Mais**
---

**Aprendizado Supervisionado**

É o modelo mais comum. Treinamos o algoritmo com um volume inicial de dados, mostrando características e o resultado obtido. O processo de aprendizado leva a prever a saída a partir de novos dados de entrada.

**Aprendizado Não Supervisionada**

É o modelo utilizado quando não temos os dados de resultado obtido na base de dados inicial. O algoritmo tenta entender as características e gerar agrupamentos (*clusters*), muitas vezes para depois passar cada *cluster* por um processo de aprendizado supervisionado.

**Aprendizado Por Reforço**

É o modelo mais indicado para tomada de decisões. Funciona por tentativa e erro, onde a máquina é exposta a vários processos decisórios e vai aprendendo conforme as decisões tomadas forem acertadas ou não.

**Redes Neurais e** ***Deep Learning***

Trabalha com sucessivas camadas de aprendizado para melhorar o processo de tomada de decisão. É utilizado para descobrir padrões em dados desestruturados, ideal para reconhecimento de imagens e de fala, por exemplo.

Normalmente as camadas são formadas por combinações de algoritmos de aprendizado supervisionado e não supervisionado.

#### **Desafio - Decida quanto você vai escolher para treino e quanto para teste? Quanto as pessoas utilizam e qual o motivo?**
---

A divisão mais comum é na proporção 70%\~30% entre dados de treino e de teste. Gosto de usar essa proporção sempre que possível. No caso específico do COVID-19 tenho encontrado dificuldade para utilizar essa proporção em razão da pequena quantidade de dados disponíveis.

No módulo anterior foi utilizado quase 84% (400 das 478 linhas) dos dados para treino, justamente por causa disso.

A função **```train_test_split```** faz a divisão tendo como parâmetro *default* a proporção de 75%\~25%, mas isso pode ser alterado mexendo no parâmetro **```test_size```**, que recebe um número entre 0 e 1 que vai indicar a proporção do teste (o resto é treino).

Para fazer uma divisão entre dados de treino, de teste e de validação, encontrei a função abaixo.


````python
import pandas as pd
from sklearn.model_selection import train_test_split

def split_stratified_into_train_val_test(df_input, stratify_colname='y',
                                         frac_train=0.6, frac_val=0.15, frac_test=0.25,
                                         random_state=None):
    '''
    Splits a Pandas dataframe into three subsets (train, val, and test)
    following fractional ratios provided by the user, where each subset is
    stratified by the values in a specific column (that is, each subset has
    the same relative frequency of the values in the column). It performs this
    splitting by running train_test_split() twice.

    Parameters
    ----------
    df_input : Pandas dataframe
        Input dataframe to be split.
    stratify_colname : str
        The name of the column that will be used for stratification. Usually
        this column would be for the label.
    frac_train : float
    frac_val   : float
    frac_test  : float
        The ratios with which the dataframe will be split into train, val, and
        test data. The values should be expressed as float fractions and should
        sum to 1.0.
    random_state : int, None, or RandomStateInstance
        Value to be passed to train_test_split().

    Returns
    -------
    df_train, df_val, df_test :
        Dataframes containing the three splits.
    '''

    if frac_train + frac_val + frac_test != 1.0:
        raise ValueError('fractions %f, %f, %f do not add up to 1.0' % \
                         (frac_train, frac_val, frac_test))

    if stratify_colname not in df_input.columns:
        raise ValueError('%s is not a column in the dataframe' % (stratify_colname))

    X = df_input # Contains all columns.
    y = df_input[[stratify_colname]] # Dataframe of just the column on which to stratify.

    # Split original dataframe into train and temp dataframes.
    df_train, df_temp, y_train, y_temp = train_test_split(X,
                                                          y,
                                                          stratify=y,
                                                          test_size=(1.0 - frac_train),
                                                          random_state=random_state)

    # Split the temp dataframe into val and test dataframes.
    relative_frac_test = frac_test / (frac_val + frac_test)
    df_val, df_test, y_val, y_test = train_test_split(df_temp,
                                                      y_temp,
                                                      stratify=y_temp,
                                                      test_size=relative_frac_test,
                                                      random_state=random_state)

    assert len(df_input) == len(df_train) + len(df_val) + len(df_test)

    return df_train, df_val, df_test
````

Vamos criar um pequeno *DataFrame* com dados exemplificativos para entender o funcionamento dos diferentes métodos de preenchimento do comando **```fillna```**, que utilizaremos para preencher as lacunas em nosso *dataset*.

O método **```ffill```** preenche para frente, colocando os dados encontrados para a próxima lacuna.

E se tivermos um paciente sem leitura nenhuma?

````python
dados_fake = pd.DataFrame([[1,1],[1,None],[2, None],[2,None]], columns=['Paciente', 'Leitura'])
dados_fake
````

````python
dados_fake.fillna(method='ffill')
````

Usando o comando para o *DataFrame* inteiro, sem indicar algum tipo de quebra, o resultado foi que os dados do paciente 1 foram utilizados para o paciente 2, o que não pode acontecer.

Temos que agrupar por paciente, para impedir isso.

````python
dados_fake = pd.DataFrame([[1,1,1],[1,None,2],[2, None,1],[2,None,2]], columns=['Paciente', 'Leitura', 'Janela'])
dados_fake
````

````python
dados_fake.groupby('Paciente').fillna(method='ffill')
````

````python
dados_fake.groupby('Paciente', as_index=False).fillna(method='ffill')
````

````python
dados_fake.groupby('Paciente', as_index=True).fillna(method='ffill')
````

Nesse caso podemos ver que o **Python** preencheu corretamente os dados do paciente 1 e, como não encontrou nenhum dado para o paciente 2, ele permaneceu com os dados nulos.

O Guilherme disse para colocar o **`as_index`** como False, mas não notei nenhuma alteração no resultado... postei dúvida no **Discord**, mas ainda não tive nenhuma resposta.

````python
dados_fake = pd.DataFrame([[1,None,1],[1,2.5,2],
                           [2, 3.0,1],[2,None,2]],
                           columns=['Paciente', 'Leitura', 'Janela'])
dados_fake
````

````python
dados_fake.groupby('Paciente', as_index=False).fillna(method='ffill')
````

Nesse exemplo podemos ver que o valor do paciente 2 foi preenchido para frente, mas o paciente 1, que tinha a primeira medição em branco, permaneceu em branco.

Agora precisamos mandar preencher para frente e para trás. Vamos repetir o comando acima, agregando um *back fill* na sequência.

````python
dados_fake = pd.DataFrame([[1,None,1],[1,2.5,2],
                           [2, 3.0,1],[2,None,2]],
                           columns=['Paciente', 'Leitura', 'Janela'])
dados_fake
````

Como podemos ver, foi feito o preenchimento das informações que estavam nas colunas de dados contínuos. Mas esse comando não altera diretamente a base de dados, ele simplesmente nos mostra como ficaria.

Para que o resultado passe a ser a nossa base de dados, precisamos atribuir o resultado para o *DataFrame*.

Para fazer isso e já colocar na base de dados os demais dados categóricos, vamos montar uma função específica para fazer todo esse processo de limpeza da base de dados.

````python
def preenche_tabela(dados):
    dados_continuos_colunas = dados.iloc[:,13:-2].columns ## as colunas com dados contínuos, entre a 13 e a penúltima
    dados_continuos = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index=False)[dados_continuos_colunas].fillna(method='bfill').fillna(method='ffill')
    dados_categoricos = dados.iloc[:,:13] ## as primeiras 13 colunas
    saida = dados.iloc[:,-2:] ## as duas últimas colunas (WINDOW e ICU)
    dados_finais = pd.concat([dados_categoricos, dados_continuos, saida], ignore_index = True, axis = 1)
    dados_finais.columns = dados.columns 
    return(dados_finais)
````
